## Maximum intensity Projection (MIP) plot

Visualization of 3d categorical or numerical values with [maximum intensity projection](https://svi.nl/MaximumIntensityProjection).

![mip](https://svi.nl/dl3163?display&x=400&y=343)

```python
%matplotlib inline

import anndata
from spacipy import mip_spatial
```

```python
adata = anndata.read_h5ad('../_static/data/example.mipplot.h5ad')
```

```python
adata
```

```python
adata.obs['spc'] = 'c' + adata.obs['spc'].astype(str)
adata.obs['spc'].unique()
```

```python
adata.uns['3d_spatial'] = adata.obs[['x', 'y', 'z']].values
```

```python
color_dict = {
    'c25': '#8c3bff', 
    'c28': '#f2007b', 
    'c31': '#15e18c', 
    'c19': '#ff7ed1', 
    'c30': '#bac389', 
    'c29': '#ff7752'
    }
```

```python
mip_spatial(adata, 
            color='spc', 
            cmap=color_dict, 
            ps=1
            )
```

```python
cyan = '#00FFFF'
magenta = '#FF00FF'
colors = [cyan, magenta]
genes = ['SMESG000008973.1', 'SMESG000005043.1']

mip_spatial(adata, 
            color=genes, 
            cmap=colors, 
            ps=1
            )
```
